```python
import pandas as pd
import numpy as np


# Count the number of missing values in each column and find the maximum number of missing values.
# How to replace missing values of `multiple numeric` columns with the mean?
# Calculate the average price of different manufacturers.
# How to replace missing values of `price` columns with the mean depending on its manufacturers?


url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'

df = pd.read_csv(url)
df2 = df.copy()

columns_missing_values_counts = {}
for i in df.columns:
  columns_missing_values_counts.setdefault(i, 0)
  columns_missing_values_counts[i] = df[i].isnull().sum()

# the number of missing values in each column
print(columns_missing_values_counts)

# the maximum number of missing values.
print(max(columns_missing_values_counts.values()))

print("平均價錢 : %f", df['Price'].mean())
# replace missing values of `multiple numeric` columns with the mean
for i in df.columns:
  if df[i].dtypes == "float64":
    df[i].fillna(df[i].mean(), inplace=True)


# Calculate the average price of different manufacturers.
print(df2.groupby(by='Manufacturer')['Price'].mean())

# df['Price'].fillna(df.groupby(by='Manufacturer')['Price'].mean(), inplace=True)
df2['Price'].fillna(df2.groupby(by='Manufacturer')['Price'].transform('mean'), inplace=True)
df2
```

```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
data = pd.read_csv(url)
data1 = data.copy()
data2 = data.copy()

# Count the number of missing values in each column and find the maximum number of missing values.
print(data.isna().sum(axis = 0).max())

# How to replace missing values of multiple numeric columns with the mean?
data1 = data.fillna(data._get_numeric_data().mean())
display(data1.head())

# Calculate the average price of different manufacturers.
display(data.groupby("Manufacturer").Price.mean())

# How to replace missing values of price columns with the mean depending on its manufacturers?
data2.Price = data2.groupby('Manufacturer').Price.transform(lambda x: x.fillna(x.mean()))
display(data)
```
